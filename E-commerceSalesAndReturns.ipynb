```python
# Aggregate revenue by state

df['Revenue'] = df['price'] + df['freight_value']

revenue_by_state = df.groupby('customer_state')['Revenue'].sum().sort_values(ascending=False).astype(int)

print(revenue_by_state)

import matplotlib.pyplot as plt

revenue_by_state.plot(kind='bar', figsize=(12,6))

plt.title("Revenue by State")
plt.xlabel("State")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
# Return per month

df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])

df['Month'] = df['order_purchase_timestamp'].dt.to_period('M')

total_orders = df.groupby('Month')['order_id'].nunique()
cancelled_orders = (
    df[df['order_status'] == 'canceled']
    .groupby('Month')['order_id']
    .nunique()
)

monthly_return_rate = (cancelled_orders / total_orders * 100).fillna(0).astype(int)

print(monthly_return_rate)

# plotting

monthly_return_rate.plot(
    kind='line',
    marker='o',
    figsize=(10,5)
)

plt.title("Monthly Return (Cancellation) Rate")
plt.xlabel("Month")
plt.ylabel("Return Rate (%)")
plt.grid(True)
plt.show()
```

```python
# Return per product

total_product_orders = df.groupby(
    'product_category_name_english'
)['order_id'].nunique()

cancelled_product_orders = (
    df[df['order_status'] == 'canceled']
    .groupby('product_category_name_english')['order_id']
    .nunique()
)

# Return Rate
product_return_rate = (
    cancelled_product_orders /
    total_product_orders * 100
).fillna(0).astype(int)

print(product_return_rate.sort_values(ascending=False))
```

```python
# Return rate rank

# Total orders by product category
total_orders = df.groupby(
    'product_category_name_english'
)['order_id'].nunique()

# Cancelled orders by product category
cancelled_orders = (
    df[df['order_status'] == 'canceled']
    .groupby('product_category_name_english')['order_id']
    .nunique()
)

# Return (Cancellation) Rate
return_rate = (
    cancelled_orders /
    total_orders * 100
).fillna(0)

# Top 10 Highest Return Rate
top_return_rate = return_rate.sort_values(
    ascending=False
).head(10)

print(top_return_rate)

plt.figure(figsize=(12,6))

top_return_rate.plot(
    kind='barh'
)

plt.title("Top 10 Product Categories by Return (Cancellation) Rate")
plt.xlabel("Return Rate (%)")
plt.ylabel("Product Category")
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
# # Business Report: Where Money is Made and Where it Leaks

# ## Executive Summary

# The analysis of the Olist E-Commerce dataset provides valuable insights into the company's sales performance, customer purchasing behavior, and operational efficiency. Revenue is primarily generated through high-demand product categories, repeat customers, and sales from major Brazilian states. However, revenue leakage occurs due to order cancellations, freight costs, and lower-performing product categories.

# ### Where Money is Made

# * The highest revenue comes from the top-selling product categories, indicating strong customer demand.
# * Customers from major states such as São Paulo (SP), Rio de Janeiro (RJ), and Minas Gerais (MG) contribute the largest share of total sales.
# * Most revenue is generated during peak shopping months, showing seasonal demand patterns.
# * Digital payment methods, especially credit cards, account for the majority of successful transactions.
# * A small group of loyal customers and high-performing sellers contribute a significant portion of total revenue.

# ### Where Money Leaks

# * Cancelled orders reduce potential revenue and negatively affect customer satisfaction.
# * High freight charges increase operational costs and reduce overall profit margins.
# * Some product categories have relatively high cancellation rates, indicating issues with inventory, delivery, or customer expectations.
# * Low-performing product categories generate limited revenue while still consuming storage and logistics resources.
# * Delivery delays may contribute to negative customer experiences and lower review scores.

# ## Recommendations

# 1. Reduce order cancellations by improving inventory management and delivery accuracy.
# 2. Focus marketing campaigns on high-performing product categories and high-revenue states.
# 3. Optimize shipping and logistics to reduce freight costs.
# 4. Strengthen customer retention through loyalty programs and personalized offers.
# 5. Improve the quality and availability of products with higher cancellation rates.

# ## Conclusion

# The analysis shows that the company's revenue is driven by popular product categories, loyal customers, and strong regional markets. At the same time, cancellations and logistics-related costs are the primary sources of revenue leakage. By addressing these operational issues and focusing on profitable segments, the business can improve customer satisfaction, increase profitability, and achieve sustainable growth.
```
